```python
import pandas as pd
```

```python
def get_audio_length(filename):
    import wave
    import contextlib
    with contextlib.closing(wave.open(filename,'r')) as f:
        frames = f.getnframes()
        rate = f.getframerate()
        duration = frames / float(rate)
        return duration
```

```python
import hparams
root = '/home/lokhiufung/data/'
lang = hparams.LANG
dataset = hparams.DATASET_NAME

meta_data = root + '/' + lang + '/' + 'src/' + dataset + '/' + 'meta_data.csv'
```

```python
df = pd.read_csv(meta_data, header=0)
df = df[df['is_available'] == 1]
df.head()
```

```python
# get length column
df['duration'] = df['wav_filename'].apply(get_audio_length)
df = df.sort_values(by=['duration'])
df.head()
```

```python
# remove duration == 0
# update is_available to 0
# def update_func(duration, is_available):
#     is_ava = is_available
#     if is_available == 1:
#         is_ava = is_available if duration > 1e-6 else 0
#     return is_ava
df['is_available'] = [is_available if duration > 1e-6 else 0 for is_available, duration in df[['is_available', 'duration']].values]
df.head()
```

```python
# export updated csv
df.to_csv(meta_data, index=False)
```
